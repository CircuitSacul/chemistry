# Notes

Use `reaction_from_string` to convert a string to a `Reaction` object. Use the following syntax:<br>
 - A single reactant/product should look like this: `<moles><name>[^<order>][_eq] (<state>)`<br>
 - A reaction should be a list of reactants, seperated by ` + ` followed by an arrow (`->`) and a list of products seperated by ` + `.<br>
 - For example, a valid reaction string would be `CH4^1_eq + 2O2^2_eq -> CO2_eq + 2H2O_eq`<br>

# Examples

```python
# Convert a string to a Reaction object
r = reaction_from_string("CH4^1_eq (g) + 2O2^2_eq (g) -> CO2_eq (g) + 2H2O_eq (l)")
```

```python
# Print the __repr__
r
```
