# Notes

Equilibrium Constant:<br>
For the reaction `aA + bB -> cC + dD`<br>
`K = ([C]_eq^c * [D]_eq^d) / ([A]_eq^a * [B]_eq^b)`

Notes:
 - `[X]` Means concentration of X
 - `X_x` Means X subscript x
 - `X^x` Means X superscript x
 - `eq` means at equilibrium
 - `K` is the equilibrium constant

# Examples

```python
# Determine K from a reaction
r = Reaction(
    reactants=[
        Item(name="NO", state=ItemState.G, moles=2, at_eq=True, concentration=0.024),
        Item(name="H2", state=ItemState.G, moles=2, at_eq=True, concentration=0.024),
    ],
    products=[
        Item(name="N2", state=ItemState.G, moles=1, at_eq=True, concentration=0.038),
        Item(name="H2O", state=ItemState.G, moles=2, at_eq=True, concentration=0.076),
    ]
)
print(r.k, r.k_unit)
```

```python
# Get the equilibrium equation
r = Reaction(
    reactants=[
        Item(name="CaCO3", state=ItemState.G, moles=1),
    ],
    products=[
        Item(name="N2", state=ItemState.S, moles=1),
        Item(name="CO2", state=ItemState.G, moles=1),
    ]
)
print(r.k_equation)
```

```python
r = Reaction(
    reactants=[
        Item(name="CS", state=ItemState.G, moles=1, at_eq=True),
        Item(name="H2", state=ItemState.G, moles=4, concentration=0.115, at_eq=True),
    ],
    products=[
        Item(name="CH4", state=ItemState.G, moles=1, concentration=0.061, at_eq=True),
        Item(name="H2S", state=ItemState.G, moles=2, concentration=0.045, at_eq=True),
    ],
    k=27.8
)
r.solve_missing_concentrations()
```
